```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime,date
from plotly import express as px
import re
from plotly.subplots import make_subplots
```

```python
import zipfile
import os
local_zip = '/content/drive/MyDrive/Colab Notebooks/archive (2).zip'
zip_ref = zipfile.ZipFile(local_zip, 'r')
os.mkdir('/content/DataProfSal')
zip_ref.extractall('/content/DataProfSal')
zip_ref.close()
```

```python
path = '/content/DataProfSal/Salary Dataset.csv'
df = pd.read_csv(path)
df.info()
df.head()
```

```python
backup=df.copy()
```

```python
sal=df['Salary']
sal=list(sal.values)
numeric_string=[]
for i in sal:
  numeric_string.append(int(re.sub("[^0-9]", "", i)))
numeric_string[:10]
```

```python
df.drop(columns=['Salary'], inplace=True)
```

```python
df['Salary']=numeric_string
```

```python
df.head()
```

```python
top_sal = df.sort_values(by=['Salary'], ascending=False).head(20)
fig=px.bar(data_frame=top_sal[['Salary', 'Company Name']], x='Company Name', y='Salary', title="Top 20 Salaries Offered")
fig.show()
```

```python
df.loc[df['Company Name']=="Google"].sort_values(by=['Salary'], ascending=False).head()
```

```python
location_wise = df[['Salary','Location']].groupby("Location").mean()
```

```python
location_wise
```

```python
fig=px.bar(data_frame=location_wise, title="Location Wise Average sal")
fig.show()
```

```python
company_wise = df[['Salary','Company Name']].groupby("Company Name").mean().sort_values(by=['Salary'], ascending=False).head(10)
company_wise
```

```python
fig=px.bar(data_frame=company_wise, title="Top 10 Avearge Sal offering Companies")
fig.show()
```

```python
temp=df.loc[df['Location']=="Bangalore"][['Location','Job Title']].groupby('Job Title').count().reset_index().values
```

```python
for i,lo in enumerate(l):
  temp=df.loc[df['Location']==lo][['Location','Job Title']].groupby('Job Title').count().reset_index().values
  d=pd.DataFrame(temp, columns=["Job Title", "Count"])
  f=px.pie(data_frame=d, values='Count', names='Job Title', labels='Job Title', title=lo)
  f.show()
  del d
```
