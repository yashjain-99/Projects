```python
import pandas as pd
Training=pd.read_csv('/content/final_train.csv')
Testing=pd.read_csv('/content/final_test.csv')
```

```python
train_dataset = Training.sample(frac=0.8, random_state=0)
test_dataset = Training.drop(train_dataset.index)
```

```python
train_features = train_dataset.copy()
test_features = test_dataset.copy()

train_labels = train_features.pop('Price')
test_labels = test_features.pop('Price')
```

```python
train_features.head()
```

```python
train_features.to_csv('/content/train_features.csv', index=False)
```

```python
train_features.describe()
```

```python
train_stats = train_features.describe()
train_stats = train_stats.transpose()
train_stats
```

```python
def norm(x):
  return (x - train_stats['mean']) / train_stats['std']
  
normed_train_data = norm(train_features)
normed_test_data = norm(test_features)
normed_train_data.head()
```

```python
import tensorflow as tf
model = tf.keras.Sequential([
    tf.keras.layers.Dense(256, activation='relu', input_shape=[len(train_features.keys())]),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(64, activation='relu'),
    tf.keras.layers.Dense(1)
  ])
model.summary()
```

```python
optimizer = tf.keras.optimizers.RMSprop(0.001)
model.compile(loss='mse',
                optimizer=optimizer,
                metrics=['mae', 'mse'])
```

```python
import tensorflow_docs as tfdocs
import tensorflow_docs.modeling 
EPOCHS = 3000

history = model.fit(
  normed_train_data, train_labels,
  epochs=EPOCHS, validation_split = 0.2, verbose=0,
  callbacks=[tfdocs.modeling.EpochDots()])
```
